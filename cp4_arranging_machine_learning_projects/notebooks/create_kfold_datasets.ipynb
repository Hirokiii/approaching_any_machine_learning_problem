```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn import datasets
from sklearn import manifold

%matplotlib inline
```

```python
# read datasets: MNIST
data = datasets.fetch_openml(
    "mnist_784",
    version=1,
    return_X_y=True
)
```

```python
pixel, labels = data
pixel.head()
```

```python
df = pixel
df["label"] = labels
df.head()
```

```python
from sklearn import model_selection

if __name__ == "__main__":
    df_here = df.copy()
    
    df_here["kfold"] = -1
    
    df_here = df_here.sample(frac=1).reset_index(drop=True)
    
    kf = model_selection.KFold(n_splits=5)
    
    for fold, (trn_, val_) in enumerate(kf.split(X=df_here)):
        df_here.loc[val_, "kfold"] = fold
        
df_here.head()
```

```python
# shuffling the df
df = df_here.sample(frac=1).reset_index(drop=True)

train_len = int(len(df) * 0.7)
test_len = len(df) - train_len


# top 1000 rows are selected for training
df_train = df.head(train_len)

# bottom 599 values for testing/validation
df_test = df.tail(test_len)
```
